```python
import pandas as pd
df_txt=pd.read_csv('essential-word-list.txt', header=None)
stopwords_list=[]
for word in df_txt[0]:
    if pd.notna(word):
        clean_word=str(word).strip().lower()
        stopwords_list=stopwords_list+[clean_word]
#print(stopwords_list)
df = pd.read_csv('dataset_with_abstract.csv')
def cal_lexical_density(text):
    if pd.isna(text):
        return 0.0
    words=str(text).lower().split()
    total_words=len(words)
    if total_words==0:
        return 0.0
    scount1 = 0
    for w in words:
        if w in stopwords_list:
            scount1 = scount1 + 1
    content_words = total_words - scount1
    lexical_density = content_words / total_words
    return lexical_density
df['Lexical_Density'] = df['abstract'].apply(cal_lexical_density)
df.to_csv('dataset_with_vocab.csv', index=False)
df.head()

```
