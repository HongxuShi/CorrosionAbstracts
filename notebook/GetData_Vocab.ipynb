```python
import pandas as pd
df_txt=pd.read_csv('essential-word-list.txt', header=None)
stopwords_list=[]
for word in df_txt[0]:
    if pd.notna(word):
        clean_word=str(word).strip().lower()
        stopwords_list=stopwords_list+[clean_word]
#print(stopwords_list)
df = pd.read_csv('dataset_with_abstract.csv')
def fun2(text):
    words=str(text).lower()
    total_words=words.count(' ')+1
    newtext=" " + words + " "
    scount1 = 0
    for w in stopwords_list:
        w_count=newtext.count(" " + w + " ")
        scount1=scount1 + w_count
    content_words = total_words - scount1
    if total_words > 0:
        lexical_density = content_words / total_words
    else:
        lexical_density = 0
    return pd.Series([total_words, scount1, lexical_density])
new_cols = ['Total_Words', 'Stop_Words_Count', 'Lexical_Density']
df[new_cols] = df['abstract'].apply(fun2)
print(df[['title', 'Total_Words', 'Stop_Words_Count', 'Lexical_Density']].head())
df.to_csv('dataset_with_vocab_features.csv', index=False)
df.head()

```
