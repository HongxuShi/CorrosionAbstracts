```python
import pandas as pd
import numpy as np
```

```python
#以下几个cell是提取hedge数据的实现，由石洪旭完成
df = pd.read_csv('dataset_with_abstract.csv')
print(df.head())
```

```python
df_abs = df["abstract"]
df_hedge = pd.read_csv('hedge_data.txt',header=None)
print(df_abs.head(),"\n",df_hedge.head())
```

```python
def count_hedges(text):
  if pd.isna(text):
      return 0
  text = text.lower()
  return sum(df_hedge[0].str.lower().apply(lambda h: h in text))

df['hedge'] = df['abstract'].apply(count_hedges)
print(df['hedge'].head(),"\n",df.head())
df.to_csv("dataset_with_hedge.csv")
```

```python
#Mean word length特征提取，由王淳完成
from pandas import DataFrame
df=pd.read_csv('dataset_with_abstract.csv')
#print(df.head())
def fun1(text):
    ccount_1 = len(text)
    pun=',.\'"!-%~'
    punc_count = 0
    text_nopunc = ""
    for char in text:
        if char in pun:
            punc_count+=1
        else:
            text_nopunc+=char
    ccount_2=len(text_nopunc)
    space_c=text_nopunc.count(' ')
    word_c=space_c+1
    if word_c>0:
        mean_word_length = ccount_2/word_c
    else:
        mean_word_length = 0
    return pd.Series([ccount_1, punc_count, ccount_2, space_c, word_c, mean_word_length])
new_cols = ['Original_CCount', 'PunCount', 'CCount_No_Punc', 'Space_Count', 'Word_Count', 'Mean_Word_Length']
df[new_cols] = df['abstract'].apply(fun1)
print(df[['title', 'Word_Count', 'Mean_Word_Length']].head())
dataset_with_features_Meanwordlength='dataset_with_features.csv'
df.to_csv(dataset_with_features_Meanwordlength, index=False)
```

```python
#以下几个cell是提取jargon数据的实现，由石洪旭完成
df_jargon = pd.read_csv("df_j.txt", sep="  ", engine='python', header=None)
df_jargon = df_jargon[1]
df_jargon = df_jargon.dropna()
print(df_jargon)
```

```python
def count_jargons(text):
  if pd.isna(text):
      return 0
  text = text.lower()
  return sum(df_jargon.str.lower().apply(lambda j: j in text))

df['jargon'] = df['abstract'].apply(count_jargons)
print(df['jargon'].head(),"\n",df.head())
df.to_csv("dataset_with_jargon.csv")
```

```python
#Lexical density特征提取，由王淳完成
df_txt=pd.read_csv('essential-word-list.txt', header=None)
stopwords_list=[]
for word in df_txt[0]:
    if pd.notna(word):
        clean_word=str(word).strip().lower()
        stopwords_list=stopwords_list+[clean_word]
#print(stopwords_list)
df = pd.read_csv('dataset_with_abstract.csv')
def fun2(text):
    words=str(text).lower()
    total_words=words.count(' ')+1
    newtext=" " + words + " "
    scount1 = 0
    for w in stopwords_list:
        w_count=newtext.count(" " + w + " ")
        scount1=scount1 + w_count
    content_words = total_words - scount1
    if total_words > 0:
        lexical_density = content_words / total_words
    else:
        lexical_density = 0
    return pd.Series([total_words, scount1, lexical_density])
new_cols = ['Total_Words', 'Stop_Words_Count', 'Lexical_Density']
df[new_cols] = df['abstract'].apply(fun2)
print(df[['title', 'Total_Words', 'Stop_Words_Count', 'Lexical_Density']].head())
df.to_csv('dataset_with_vocab_features.csv', index=False)
df.head()

```

```python
#Lexical richness特征提取，由王淳完成
from pandas import DataFrame
df=pd.read_csv('dataset_with_abstract.csv')
def fun4(text):
    words= str(text).strip().lower()
    total_words = words.count(' ') + 1
    newtext=" " + words + " "
    word_count = 0
    seen_words = " "
    current_word = ""
    for char in newtext:
        if char != ' ' and char not in ',.\'"!-%~':
            current_word = current_word + char
        else:
            if current_word not in seen_words:
                word_count = word_count + 1
                seen_words = seen_words + " " + current_word + " "
            current_word = ""
    if current_word != "":
        if (" " + current_word + " ") not in seen_words:
            word_count=word_count + 1
    if total_words > 0:
        lexical_richness = word_count / total_words
    else:        
        lexical_richness = 0
    return pd.Series([word_count, total_words, lexical_richness])
new_cols = ['Unique_Words_Type', 'Total_Words_Token', 'Lexical_Richness_TTR']
df[new_cols] = df['abstract'].apply(fun4)
print(df[['title', 'Unique_Words_Type', 'Total_Words_Token', 'Lexical_Richness_TTR']].head())
df.to_csv('dataset_with_complexity.csv', index=False)
```

```python
#Average sentence length特征提取，由王淳完成
df=pd.read_csv('dataset_with_abstract.csv')
def fun3(text):
    text_str = str(text)
    sentence_count = text_str.count('.')
    word_count=text_str.count(' ')+1
    avg_sentence_length = word_count / sentence_count
    return pd.Series([sentence_count, word_count, avg_sentence_length])
new_cols = ['Sentence_Count', 'Word_Count', 'Avg_Sentence_Length']
df[new_cols] = df['abstract'].apply(fun3)
print(df[['title', 'Sentence_Count', 'Word_Count', 'Avg_Sentence_Length']].head())
df.to_csv('dataset_with_abstract_features.csv', index=False)
```

```python
#以下几个cell实现对hedge和hargon数据进行密度化，由石洪旭完成
df = pd.read_csv("dataset_with_jargon.csv")
df['jargon_density'] = df["jargon"] / ( df['abstract'].str.count(" ") + 1 )
print(df['jargon_density'].head(),"\n",df.head())
df.to_csv("dataset_with_JD.csv")
```

```python
df = pd.read_csv("dataset_with_hedge.csv")
df['hedeg_density'] = df["hedge"] / ( df['abstract'].str.count(" ") + 1 )
print(df['hedeg_density'].head(),"\n",df.head())
df.to_csv("dataset_with_HD.csv")
```
